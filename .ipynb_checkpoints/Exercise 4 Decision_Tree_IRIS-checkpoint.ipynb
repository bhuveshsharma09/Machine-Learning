# Other Classification Exercise - Decision Tree Model



Build and validate a decision tree classification model with the following dataset.

Input dataset: 'iris-data-clean.csv'

Features: 'sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm'

Label: 'class'

### <font color='red'>1) Loading the data and calling required libs
 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, confusion_matrix,roc_curve,auc

from sklearn import tree
import graphviz
from graphviz import Source
from dtreeviz.trees import *

from sklearn.model_selection import cross_val_score

from IPython.display import display, Markdown
import warnings
warnings.filterwarnings("ignore")



%matplotlib inline
```

```python
data = pd.read_csv('iris-data-clean.csv')
data.head(5)
```

### <font color='red'>2) Feature Engineering
 

```python
data.info()
```

### <font color='blue'>obsertaion: the class column is object (string) it should be converted into numeric (unordered)

```python
#get all the unique class names
unique_classes = list(data['class'].unique())
```

```python
def convert_object_to_numeric(x):
    if x == unique_classes[0]:
        return 0
    if x== unique_classes[1]:
        return 1
    else:
        return 2
```

```python
# encode the numerics for the labels
data['numeric_class'] = data['class'].apply(lambda x: convert_object_to_numeric(x))
```

```python
# drop the class object column
data.drop('class',axis=1,inplace=True)
```

```python
#check final data set
# using sample() to select rows randomly
data.sample(5)
```

```python
# now need to make numeric_class column to be categorical
data['cat_class']= data['numeric_class'].astype('category')
```

```python
#data['cat_class'] = data['cat_class'].cat.codes
```

```python
data.drop('numeric_class',axis=1,inplace=True)
```

```python
data.sample(5)
```

```python
data.info()
```

### <font color='red'>4) Data spliting
 

```python
# take all rows and all columns except last one
X = data.iloc[:,:-1]

# take all rows and only last column
y = data.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
#checking if the split is balanced
data['cat_class'].value_counts()
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### <font color='red'>5) Create a model 
 

```python
model = DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=None,random_state=42)
```

```python
model.fit(X_train,y_train)
```

### <font color='red'>6) Evaluation of model
 

```python
y_pred = model.predict(X_test)
accuracy_score_model1 = accuracy_score(y_test,y_pred)
#print(accuracy_score_model1)
Markdown('Accouracy Score is <strong><font color=\'blue\'>{}</strong></font><br/>'.format(accuracy_score_model1))
```

```python
#visualising the confusion matrix 
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,annot_kws={"size":30},fmt='d',cmap='Blues')\
.set_title("Decision Tree Confusion Matric")
```

### <font color='red'>12) Finding the best hyper parameter using grid search cv
 

```python
from sklearn.model_selection import GridSearchCV

def dtree_grid_search(XX,yy,nfolds):
    #create a dictionary of all values we want to test
    param_grid = {'criterion':['gini','entropy'],'max_depth': np.arange(1, 15),'random_state':[42]}
    
    # decision tree model
    dtree_model=DecisionTreeClassifier()
    
    #use gridsearch to test all values
    dtree_gscv = GridSearchCV(dtree_model, param_grid, cv=nfolds,scoring='accuracy')
    
    #fit model to data
    dtree_gscv.fit(XX, yy)
   # print(dtree_gscv.best_estimator_)

    print(dtree_gscv.best_score_)
    return dtree_gscv.best_params_



dtree_grid_search(X_train,y_train,5)
```

```python
final_model = DecisionTreeClassifier(criterion='gini',max_depth=2,random_state=42)
final_model.fit(X_train,y_train)

y_pred_final_model = final_model.predict(X_test)
accuracy_score_final_model = accuracy_score(y_test,y_pred_final_model)
print(accuracy_score_final_model)

scores = cross_val_score(final_model, X_train, y_train, cv=5)
print(scores)
```
